```python
pop = 'MSO'
subjects = ['1','2', '3', '4','5']
sounds = ['tone_0.1kHz', 'tone_1.kHz', 'tone_10.kHz', 'whitenoise']
v_maxes = [None, None, None, None]
```

```python
for f, v in zip(sounds, v_maxes):
    print(f, '\n')
    for sub in subjects:
        print('subject', sub, '\n')
        title = f'{f}_70dB&TanCarney&subject_{sub}.pic'
        result_path = result_dir + title
        with open(result_path, "rb") as file:
            result = dill.load(file, ignore=True)
            
        fig = plot_tonotopic_heatmaps(
            data=result,
            pop=pop,
            num_cells_per_interval=50,
            title=None,
            figsize=(7, 5),
            show_sides=False,
            f_ticks=[100,1000,10000],
            vmax_given=v
        )
        plt.show()
        # Free memory
        del result
        gc.collect()
        # Free memory
```
